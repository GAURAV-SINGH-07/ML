```python
import numpy as np
import tensorflow as tf
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Load dataset from directory
dataset_dir = "/content/dataset"  # Change this to your dataset path

# Load training and validation datasets
train_dataset = image_dataset_from_directory(
    dataset_dir,
    image_size=(32, 32),  # Resizing to match the model input shape
    batch_size=32,
    label_mode='int',
    subset="training",
    validation_split=0.2,
    seed=123,
)

val_dataset = image_dataset_from_directory(
    dataset_dir,
    image_size=(32, 32),  # Resizing to match the model input shape
    batch_size=32,
    label_mode='int',
    subset="validation",
    validation_split=0.2,
    seed=123,
)

# Get the class names from the training dataset
class_names = train_dataset.class_names

# Normalize pixel values
normalization_layer = layers.Rescaling(1./255)
train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
val_dataset = val_dataset.map(lambda x, y: (normalization_layer(x), y))



def create_cnn_model():
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    # Use the class_names variable obtained earlier
    model.add(layers.Dense(len(class_names)))  # Adjust the output layer for the number of classes

    return model

# Define fuzzy decision function (same as your original code)
def fuzzy_decision(probabilities):
    probabilities = np.array(probabilities)
    low = fuzz.trimf(probabilities, [0, 0, 0.5])
    medium = fuzz.trimf(probabilities, [0, 0.5, 1])
    high = fuzz.trimf(probabilities, [0.5, 1, 1])
    fuzzy_sets = np.vstack((low, medium, high))
    decision = fuzz.defuzz(np.array([0, 1, 2]), np.max(fuzzy_sets, axis=1), 'centroid')

    if decision < 1:
        return "Low Confidence"
    elif decision < 2:
        return "Medium Confidence"
    else:
        return "High Confidence"

# Create CNN model
cnn_model = create_cnn_model()
cnn_model.compile(optimizer='adam',
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])

# Train the CNN model
cnn_model.fit(train_dataset, epochs=10, validation_data=val_dataset)

# Get the prediction probabilities from the CNN
probability_model = models.Sequential([cnn_model, layers.Activation('sigmoid')]) # Use layers.Activation
predictions = []

for images, labels in val_dataset:
    preds = probability_model.predict(images)
    predictions.extend(preds)

# Applying fuzzy decision-making on the first validation image
first_image_probs = predictions[0]
print("Class Probabilities:", first_image_probs)

fuzzy_result=fuzzy_decision(first_image_probs)
fuzzy_result
```

```python
for images, labels in val_dataset.take(1):
    plt.imshow(images[0].numpy())
    plt.title(f"Fuzzy Decision: {fuzzy_result}")
    plt.show()
```

```python
for i, (images, labels) in enumerate(val_dataset):
    for j in range(len(images)):
        first_image_probs = predictions[i * 32 + j]
        fuzzy_result = fuzzy_decision(first_image_probs)
        if fuzzy_result == "High Confidence":
            print(f"Found a high confidence image at index {i*32+j}")
            plt.imshow(images[j].numpy())
            plt.title(f"Fuzzy Decision: {fuzzy_result}")
            plt.show()
            break
```

```python
for i, (images, labels) in enumerate(val_dataset):
    for j in range(len(images)):
        first_image_probs = predictions[i * 32 + j]
        fuzzy_result = fuzzy_decision(first_image_probs)

        # Print the class probabilities and fuzzy decision for each image
        print(f"Image Index {i*32+j}")
        print("Class Probabilities:", first_image_probs)
        print("Fuzzy Decision:", fuzzy_result)
        print("-" * 50)

        # Display the image if it has High Confidence
        if fuzzy_result == "High Confidence":
            print(f"Found a high confidence image at index {i*32+j}")

            # Display the image
            plt.figure()
            plt.imshow(images[j].numpy())
            plt.title(f"Fuzzy Decision: {fuzzy_result}")
            plt.show()

            # Break after finding the first high confidence image
            break
```
